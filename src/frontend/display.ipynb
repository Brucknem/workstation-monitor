```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource, LinearAxis, Range1d

from datetime import datetime
output_notebook()

datetime_format = '%Y/%m/%d %H:%M:%S.%f'

readable_keys = [
    'timestamp',
    'memory.used [MiB]',
    'memory.free [MiB]',
    'utilization.memory [%]',
    'fan.speed [%]',
    'temperature.gpu',
    'utilization.gpu [%]',
]

display(memory_log[readable_keys])

source = ColumnDataSource(data=dict(
    timestamp = [datetime.strptime(d, datetime_format) for d in memory_log[readable_keys[0]]],
    memory_used = [int(yyy) for yy in memory_log[readable_keys[1]] for yyy in yy.split() if yyy.isdigit()],
    memory_free = [int(yyy) for yy in memory_log[readable_keys[2]] for yyy in yy.split() if yyy.isdigit()],
    utilization_memory = [int(yyy) for yy in memory_log[readable_keys[3]] for yyy in yy.split() if yyy.isdigit()],
    fan_speed = [int(yyy) for yy in memory_log[readable_keys[4]] for yyy in yy.split() if yyy.isdigit()],
    temperature_gpu = [yy for yy in memory_log[readable_keys[5]]],
    utilization_gpu = [int(yyy) for yy in memory_log[readable_keys[6]] for yyy in yy.split() if yyy.isdigit()],
))
keys = list(source.data.keys())
```

```python
# create a new plot with a title and axis labels
p = figure(title=test_log, x_axis_label=keys[0], y_axis_label=readable_keys[1], x_axis_type='datetime')

# Add a stacked area for the memory usage
p.varea_stack([keys[1], keys[2]], x=keys[0], color=("grey", "lightgrey"), source=source)
# add a line renderer with legend and line thickness
p.line(x=keys[0], y=keys[1], legend_label=readable_keys[1], line_width=2, source=source)

# show the results
show(p)
```

```python
# create a new plot with a title and axis labels
p = figure(title=test_log, x_axis_label=keys[0], y_axis_label='Usage', x_axis_type='datetime')

# add a line renderer with legend and line thickness
p.line(x=keys[0], y=keys[6], legend_label=readable_keys[6], line_width=0.5, source=source, color='green')
p.line(x=keys[0], y=keys[3], legend_label=readable_keys[3], line_width=2, source=source, color='yellow')
p.line(x=keys[0], y=keys[4], legend_label=readable_keys[4], line_width=2, source=source, color='red')
p.line(x=keys[0], y=keys[5], legend_label=readable_keys[5], line_width=2, source=source, color='blue')

# show the results
show(p)
```
